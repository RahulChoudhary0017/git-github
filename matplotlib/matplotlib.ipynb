Pie

```python
x =np.array([1,2,3,4,5])
plt.pie(x)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
label = np.array(["o2", "co2", "n", "other", "argon"])  # Added a fifth label
plt.pie(x, labels=label, startangle=90)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
label = np.array(["o2", "co2", "n", "other", "argon"])
myexplode = [0, 0.3, 0, 0, 0]
plt.pie(x, labels=label, startangle=90, explode=myexplode, shadow=True)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
label = np.array(["o2", "co2", "n", "other", "argon"])
myexplode = [0, 0.3, 0, 0, 0]
colors = ['red', 'green', 'yellow', 'blue', 'cyan']  # 5 colors for 5 slices
plt.pie(x, labels=label, startangle=90, explode=myexplode, shadow=True, colors=colors)
plt.show()
```
